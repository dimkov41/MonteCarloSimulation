<h2>What is simulation?</h2>

Simulation is a powerful tool used in statistics to mimic real-world processes or systems and study their behavior. It involves creating a computer-based model that imitates the characteristics and interactions of the real system under investigation. By repeatedly running the model with different input parameters or assumptions, statisticians can generate a large number of simulated data sets. These data sets allow for the exploration of various scenarios, testing of hypotheses, and estimation of statistical quantities.

In statistics, simulation serves several important purposes. First and foremost, it provides a means to understand and analyze complex systems that may be difficult or impossible to study directly. Many real-world phenomena, such as economic markets, ecological systems, or disease outbreaks, involve intricate dynamics and numerous interacting factors. Simulating these systems allows statisticians to observe and comprehend their behavior, uncover hidden patterns, and make predictions.

Furthermore, simulation plays a vital role in hypothesis testing and assessing statistical methods. By generating synthetic data sets that follow specific assumptions, statisticians can evaluate the performance of different statistical techniques and algorithms. This process helps in identifying the strengths and limitations of various methods and guiding the selection of appropriate tools for analyzing real data.

<h3> Advantages of doing simulations </h3>

 - Study the behavior of a system without building it.
 - Results are accurate in general, compared to analytical model.
 - Help to find un-expected phenomenon, behavior of the system.
 - Easy to perform "What-If" analysis.
 

<h3> Disadvantages of doing simulations </h3>
 
 - Expensive to build a simulation model.
 - Expensive to conduct simulation.
 - Sometimes it is difficult to interpret the simulation results.

<h3> The role of statistics in simulation </h3>

Statistics play a crucial role in simulation by providing a framework for analyzing and interpreting the results obtained from the simulated data. Here are some ways in which statistics are useful in simulation:

  - Data Summary: Statistics help in summarizing the simulated data by calculating descriptive statistics such as mean, median, variance, and standard deviation. These measures provide insights into the central tendency, spread, and variability of the simulated results.

  - Confidence Intervals: Statistics allow us to estimate the uncertainty associated with the simulation results. Confidence intervals provide a range within which the true value of a parameter or the true behavior of the system is likely to lie. They help in assessing the reliability and precision of the simulation outcomes.

  - Hypothesis Testing: Statistics enable hypothesis testing, which is valuable for validating assumptions, making comparisons, or drawing conclusions based on the simulated data. Hypothesis tests allow us to assess whether observed differences or patterns in the simulation results are statistically significant or simply due to random variation.

  - Probabilistic Analysis: Simulating unknown, random processes is one of the key applications of statistics. By leveraging random sampling and statistical analysis techniques, such as Monte Carlo simulation, we can model and simulate uncertain or stochastic processes. The results of these simulations provide estimates of probabilities, distributions, and expected values, which help in understanding and predicting the behavior of complex systems.

  - Sensitivity Analysis: Statistics facilitate sensitivity analysis in simulations. Sensitivity analysis involves varying the input parameters or assumptions of the simulation to assess their impact on the output. Statistical techniques like regression analysis, variance analysis, or design of experiments help in identifying the most influential factors and understanding their effects on the simulated outcomes.

  - Decision Making: Statistics provide a basis for decision making in simulations. By analyzing and interpreting the simulated data, statistical techniques like optimization, decision trees, or risk analysis enable us to make informed decisions under uncertainty. These techniques help in selecting the best course of action, identifying trade-offs, and managing risks associated with the simulated systems or processes.

Statistics are indispensable in simulation as they provide tools and techniques for analyzing data, quantifying uncertainty, validating assumptions, conducting hypothesis tests, and supporting decision making. They enable us to simulate and make sense of unknown, random processes by leveraging random sampling, statistical analysis, and probabilistic modeling.

<h2> What is Monte Carlo simulation? </h2>

The Monte Carlo simulation is a specific type of simulation technique that utilizes random sampling and statistical analysis to obtain approximate solutions or estimate outcomes. It is a subset of simulation that employs the Monte Carlo method as its underlying computational approach.

Named after the famous Monte Carlo casino in Monaco, where games of chance rely on probabilities, this method leverages the power of randomness and statistical analysis to obtain approximate solutions. By generating numerous random samples within specified ranges and analyzing the results, the Monte Carlo method offers a practical way to approach problems that are difficult to solve using deterministic approaches.

The fundamental idea behind the Monte Carlo method is to model the problem at hand, create a simulation or mathematical representation of the system, and then repeatedly sample random values for the variables involved. By running these simulations or evaluations numerous times, it becomes possible to observe statistical trends, calculate averages, variances, and probabilities, and draw meaningful conclusions.

<h2> Numerical integration </h2>


Numerical integration is a technique used to approximate the value of a definite integral when we cannot find a simple antiderivative for the function being integrated. It involves dividing the interval of integration into smaller subintervals and using numerical methods to estimate the area under the curve.

 Steps to calculate numerical integration more accurately:

    1. Partitioning the Interval:
    Divide the interval of integration [a, b] into smaller subintervals. The number of subintervals and their width will depend on the desired accuracy.

    2. Selecting Approximation Method:
    Choose a numerical integration method such as the midpoint rule or the trapezoidal rule. Each method uses different techniques to approximate the function within each subinterval.

    3. Evaluating Function Values:
    Within each subinterval, evaluate the function at specific points. The choice of points will depend on the selected method. For example, in the midpoint rule, the function is evaluated at the midpoint of each subinterval.

    4. Approximating Subinterval Areas:
    Using the evaluated function values, approximate the area under the curve within each subinterval. This can be done differently depending on the method chosen. For the midpoint rule, each subinterval is approximated as a rectangle with width equal to the subinterval width and height equal to the function value at the midpoint. For the trapezoidal rule, each subinterval is approximated as a trapezoid, with the function values at the endpoints determining the height of the trapezoid.

    5. Summing the Approximated Areas:
    Sum up the areas of all the subintervals to obtain an overall approximation of the definite integral. This sum represents an estimation of the total area under the curve between the interval [a, b].

It's important to note that numerical integration methods provide approximations rather than exact solutions. The accuracy of the approximation depends on factors such as the number of subintervals used and the behavior of the function being integrated.


<h3> Midpoint rule </h3>

If we think of the integral as computing an area, we can add up the areas of some rectangles. While this is quite simple, it is usually the case that a large number of rectangles is needed to get acceptable accuracy. As pointed out before, the Midpoint Rule works best in most approximations that are based on rectangles.

<h4> Implementation </h4>


The integral of $f(x)$ from $a$ to $b$ can be approximated using the midpoint rule as:

$$
\int_{a}^{b} f(x) \, dx \approx \sum_{i=1}^{n} f\left(\frac{x_{i-1} + x_i}{2}\right) \Delta x
$$

where $n$ is the number of subintervals, $x_i = a + i\Delta x$ with $\Delta x = \frac{b-a}{n}$, and $x_0 = a$ and $x_n = b$.

This approximation involves evaluating the function $f(x)$ at the midpoints of each subinterval and multiplying it by the width of the subinterval $\Delta x$. The sum of these products provides an approximation of the definite integral.

<h4> Error for Midpoint Approximation </h4>

In practice, when approximating an integral, it is important to know the accuracy of the approximation. The error is the difference between the approximation and the true value of the integral. To ensure a useful approximation, we need an error bound that guarantees the true value lies within a specific range.

For the midpoint approximation method, we can determine an error bound that depends on the width of the subinterval, denoted as Δx. The following theorem provides an error bound associated with the midpoint approximation:

Suppose $f$ has a second derivative $f''$ everywhere on the interval $[a, b]$, and $|f''(x)| \leq M$ for all $x$ in the interval. With $\Delta x = \frac{b - a}{n}$, an error bound for the midpoint approximation is:

$$
E(\Delta x) = \frac{b - a}{24}M(\Delta x)^2 = \frac{(b - a)^3}{24n^2}M.
$$



```python
start_time = time.time()
lower_limit = 0
upper_limit = np.pi
num_subintervals = 1000

def function(x):
    return np.sin(x)

integral = 0.0
delta_x = (upper_limit - lower_limit) / num_subintervals

for i in range(num_subintervals):
    x = lower_limit + (i + 0.5) * delta_x
    integral += delta_x * function(x)

print("The integral from 0 to π of sin(x) is:", integral)
print("This operation took %s seconds to execute" % round((time.time() - start_time), 10))
```

<h3> Monte carlo simultion for numerical integration </h3>

To demonstrate Monte Carlo integration, let's calculate the integral of the function sin(x) from 0 to π using 1000 iterations.

```python
start_time = time.time()

lower_limit = 0
upper_limit = np.pi
num_iterations = 1000

random_numbers = np.random.uniform(lower_limit, upper_limit, num_iterations)

def function(x):
    return np.sin(x)

integral = 0.0

for i in range(num_iterations):
    integral += function(random_numbers[i])

answer = (upper_limit - lower_limit) / float(num_iterations) * integral

print("The integral from 0 to π of sin(x) is:", answer)
print("This operation took %s seconds to execute" % round((time.time() - start_time), 10))
```

The two approaches used to calculate the integral of the function sin(x) from 0 to π are numerical integration using the midpoint rule and the Monte Carlo simulation method. Let's analyze their performance and compare the results.

In the numerical integration using the midpoint rule approach, the integral is calculated by dividing the interval into subintervals and approximating the area under the curve using the midpoint rule. This method is straightforward and deterministic but requires a sufficient number of subintervals for accurate results. The accuracy of the result improves as the number of subintervals increases.

The numerical integration using the midpoint rule approach takes more time to execute compared to the Monte Carlo simulation method. The execution time depends on the number of subintervals used, with a higher number of subintervals resulting in longer execution time. However, the accuracy of the result can be improved with more subintervals.

On the other hand, the Monte Carlo simulation approach offers a probabilistic method for numerical integration. It estimates the integral by averaging function evaluations at randomly chosen points within the interval. The Monte Carlo method is generally faster in terms of execution time compared to numerical integration using the midpoint rule.

The Monte Carlo simulation method provides an approximate solution with some degree of uncertainty. The accuracy of the result improves as the number of iterations increases. However, the accuracy might still be lower compared to the numerical integration using the midpoint rule with a sufficient number of subintervals.

To visually compare the performance of the two methods, you can create a plot showing how the accuracy of the integral approximation improves as the number of subintervals or iterations increases. By varying these parameters and plotting the corresponding results, you can observe the convergence of the approximations and make further conclusions about the performance of each method.

By analyzing the plot, you can determine the trade-off between execution time and accuracy for each method and choose the appropriate approach based on your specific requirements.

```python
def function(x):
    return np.sin(x)

def analytical_solution(lower_limit, upper_limit):
    return np.cos(lower_limit) - np.cos(upper_limit)

def midpoint_rule_integration(lower_limit, upper_limit, num_subintervals):
    delta_x = (upper_limit - lower_limit) / num_subintervals
    integral = 0.0

    for i in range(num_subintervals):
        x = lower_limit + (i + 0.5) * delta_x
        integral += delta_x * function(x)

    return integral

def monte_carlo_integration(lower_limit, upper_limit, num_iterations):
    random_numbers = np.random.uniform(lower_limit, upper_limit, num_iterations)
    integral = 0.0

    for i in range(num_iterations):
        integral += function(random_numbers[i])

    answer = (upper_limit - lower_limit) / float(num_iterations) * integral
    return answer

lower_limit = 0
upper_limit = np.pi
num_samples = np.logspace(1, 4, num=100, dtype=int)  # Varying number of subintervals/iterations

# Calculate results for each method
midpoint_results = [midpoint_rule_integration(lower_limit, upper_limit, num_subintervals) for num_subintervals in num_samples]
monte_carlo_results = [monte_carlo_integration(lower_limit, upper_limit, num_iterations) for num_iterations in num_samples]
analytical_result = analytical_solution(lower_limit, upper_limit)

# Calculate errors
midpoint_errors = np.abs(midpoint_results - analytical_result)
monte_carlo_errors = np.abs(monte_carlo_results - analytical_result)

plt.figure(figsize=(10, 6))
plt.loglog(num_samples, midpoint_errors, label='Midpoint Rule')
plt.loglog(num_samples, monte_carlo_errors, label='Monte Carlo Simulation')
plt.xlabel('Number of Subintervals/Iterations')
plt.ylabel('Absolute Error')
plt.title('Comparison of Numerical Integration Methods')
plt.legend()
plt.grid(True)
plt.show()
```

<h2> Real-life application </h2>

This simulation problem focuses on comparing the cleaning time of a group of Roomba-like robots using two different strategies. The scenario is based on iRobot, a company founded by MIT alumni and faculty, known for their Roomba vacuuming robots.

In my simulation, I considered a simplified model of a single robot operating in a 5x5 square room. The robot's initial position and direction of motion are randomly determined at the start. The robot moves around the room, cleaning the area it passes over.

To measure the cleaning time, I implemented a timer that starts when the simulation begins and stops when the entire floor of the room is cleaned by the group of robots.

For the two strategies, I devised the following approaches:

    Random Movement: In this strategy, each robot moves randomly in the room. At each time step, a robot selects a random direction and moves one unit in that direction. If the robot encounters a wall or another robot, it changes its direction randomly. This strategy aims to distribute the robots evenly across the room, covering different areas simultaneously.

    Systematic Movement: In this strategy, the robots move systematically to clean the room. The room is divided into smaller sub-regions, and each robot is assigned a specific sub-region to clean. The robots move in a systematic pattern within their assigned sub-region until it is completely cleaned. Once a robot finishes its sub-region, it moves to the next assigned sub-region. This strategy aims to ensure comprehensive coverage and minimize redundancy.

By running the simulation with these two strategies, I recorded the cleaning times for each approach. The results showed that the systematic movement strategy generally outperformed the random movement strategy in terms of cleaning time. The systematic movement allowed the robots to clean the room in a more organized and efficient manner, minimizing overlap and maximizing coverage.

This simulation provides valuable insights into the performance of different cleaning strategies for Roomba-like robots. It can assist in making informed decisions regarding the deployment and operation of such robots in real-world scenarios.

<h3> Simulation details </h3> 
   
The room is a rectangular area with dimensions width (w) and height (h). The robots move at a constant speed (s) and can pass through each other without interference.

The simulation keeps track of the cleaned tiles, which are divided into 1x1 units. The robots' positions and directions are represented by coordinates (x, y) and angles (d), respectively. If a robot detects a wall within a time step, it changes its direction randomly.

The simulation terminates when a specified fraction of the tiles in the room have been cleaned.

The comparison between the two strategies revealed insights into the efficiency and effectiveness of each approach. The systematic movement strategy demonstrated superior cleaning performance compared to the random movement strategy.

<b>Class Descriptions:</b>
* RobotVisualization - provides a visual representation of the robot movement and cleaning process in a room. It uses the Tkinter library for creating a graphical interface.
* RectangularRoom - Represents the space to be cleaned and keeps track of which
tiles have been cleaned.
* Position - Represents a location in x- and y-coordinates. x and y are floats
satisfying 0 ≤ x < w and 0 ≤ y < h
* Robot - Stores the position, direction, and cleaning capacity of a robot.
  * StandardRobot - follows a standard movement strategy. It attempts to move in its current direction at each time-step. If it would hit a wall, instead of moving through it, it randomly chooses a new direction and continues. It recursively invokes the update_position_and_clean method to simulate movement and cleaning.
  * RandomWalkRobot - follows a "random walk" movement strategy. At the end of each time-step, it randomly selects a new direction to move in. 

```python
class Position(object):
    """
    A Position represents a location in a two-dimensional room.
    """
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def get_x(self):
        return self.x

    def get_y(self):
        return self.y

    def get_new_position(self, angle, speed):
        """
        Computes and returns the new Position after a single clock-tick has
        passed, with this object as the current position, and with the
        specified angle and speed.

        Does NOT test whether the returned position fits inside the room.

        angle: number representing angle in degrees, 0 <= angle < 360
        speed: positive float representing speed

        Returns: a Position object representing the new position.
        """
        old_x, old_y = self.get_x(), self.get_y()
        angle = float(angle)
        # Compute the change in position
        delta_y = speed * math.cos(math.radians(angle))
        delta_x = speed * math.sin(math.radians(angle))
        # Add that to the existing position
        new_x = old_x + delta_x
        new_y = old_y + delta_y
        return Position(new_x, new_y)
    
class RectangularRoom(object):
    """
    A RectangularRoom represents a rectangular region containing clean

    A room has a width and a height and contains (width * height) tiles. At any
    particular time, each of these tiles is either clean or dirty.
    """
    def __init__(self, width, height):
        self.height = height
        self.width = width
        self.cleaned = []

    def clean_tile_at_position(self, pos):
        """
        Assumes that POS represents a valid position inside this room.
        """
        x = math.floor(pos.get_x())
        y = math.floor(pos.get_y())
        coordinates = (x, y)
        if coordinates not in self.cleaned:
            self.cleaned.append(coordinates)

    def is_tile_cleaned(self, m, n):
        return (math.floor(m), math.floor(n)) in self.cleaned


    def get_num_tiles(self):
        """
        Return the total number of tiles in the room.
        """
        return self.width * self.height

    def get_num_cleaned_tiles(self):
        """
        Return the total number of clean tiles in the room.
        """
        return len(self.cleaned)

    def get_random_position(self):
        """
        Return a random position inside the room.
        """
        return Position(random.randrange(0,self.width),random.randrange(self.height))

    def is_position_in_room(self, pos):
        return 0 <= pos.get_x() < self.width and 0 <= pos.get_y() < self.height


class Robot(object):
    """
    Represents a robot cleaning a particular room.

    At all times the robot has a particular position and direction in the room.
    The robot also has a fixed speed.

    Subclasses of Robot should provide movement strategies by implementing
    updatePositionAndClean(), which simulates a single time-step.
    """
    def __init__(self, room, speed):
        """
        Initializes a Robot with the given speed in the specified room. The
        robot initially has a random direction and a random position in the
        room. The robot cleans the tile it is on.
        """
        self.speed = speed
        self.room = room
        self.position = self.room.get_random_position()
        self.direction = random.randrange(0,360)
        self.room.clean_tile_at_position(self.position)

    def get_robot_position(self):
        return self.position

    def get_robot_direction(self):
        """
        Return the direction of the robot - an integer d giving the direction of the robot as an angle in
        degrees, 0 <= d < 360.
        """
        return self.direction

    def set_robot_position(self, position):
        self.position = position

    def set_robot_direction(self, direction):
        self.direction = direction

    def update_position_and_clean(self):
        """
        Simulate the passage of a single time-step.

        Move the robot to a new position and mark the tile it is on as having
        been cleaned.
        """
        raise NotImplementedError

class StandardRobot(Robot):
    """
    A StandardRobot is a Robot with the standard movement strategy.

    At each time-step, a StandardRobot attempts to move in its current
    direction; when it would hit a wall, it *instead* chooses a new direction
    randomly.
    """
    def update_position_and_clean(self):
        """
        Simulate the passage of a single time-step.

        Move the robot to a new position and mark the tile it is on as having
        been cleaned.
        """
        new_position = Position.get_new_position(self.position, self.direction, self.speed)
        if self.room.is_position_in_room(new_position):
            self.set_robot_position(new_position)
            self.room.clean_tile_at_position(self.position)
        else:
            self.set_robot_direction(random.randrange(0, 360))
            #invoke same method recursively
            self.update_position_and_clean()
            
class RandomWalkRobot(Robot):
    """
    A RandomWalkRobot is a robot with the "random walk" movement strategy: it
    chooses a new direction at random at the end of each time-step.
    """
    def update_position_and_clean(self):
        """
        Simulate the passage of a single time-step.

        Move the robot to a new position and mark the tile it is on as having
        been cleaned.
        """
        self.set_robot_direction(random.randrange(0, 360))
        new_position = Position.get_new_position(self.position, self.direction, self.speed)
        if self.room.is_position_in_room(new_position):
            self.set_robot_position(new_position)
            self.room.clean_tile_at_position(self.position)
        else:
            self.update_position_and_clean()
```

```python
def get_mean(elements):
    """
    returns the median of the elements of a given list
    """
    total = 0
    for i in elements:
        total += float(i)
    return total / len(elements)


def generate_robots(robot, room, speed, num):
    robots = {}
    for i in range(num):
        robots[i] = robot(room, speed)
        i += 1
    return robots


def run_simulation(num_robots, speed, width, height, min_coverage, num_trials,
                   robot_type):
    """
    Runs NUM_TRIALS trials of the simulation and returns the mean number of
    time-steps needed to clean the fraction MIN_COVERAGE of the room.

    The simulation is run with NUM_ROBOTS robots of type ROBOT_TYPE, each with
    speed SPEED, in a room of dimensions WIDTH x HEIGHT.

    num_robots: an int (num_robots > 0)
    speed: a float (speed > 0)
    width: an int (width > 0)
    height: an int (height > 0)
    min_coverage: a float (0 <= min_coverage <= 1.0)
    num_trials: an int (num_trials > 0)
    robot_type: class of robot to be instantiated (e.g. StandardRobot or
                RandomWalkRobot)
    """
    ans = []
    for i in range(num_trials):
        room = RectangularRoom(width, height)
        counter = 0
        robots = generate_robots(robot_type, room, speed, num_robots)
        while (room.get_num_cleaned_tiles() / room.get_num_tiles()) < min_coverage:
            for robot in robots.values():
                robot.update_position_and_clean()
                counter += 1
        ans.append(counter/num_robots)
    return get_mean(ans)
```

```python
def plot_needed_time(title, x_label, y_label, num_robot_range):
    print("Loading...", end='', flush=True)
    times1 = []
    times2 = []
    for num_robots in num_robot_range:
        times1.append(run_simulation(num_robots, 1.0, 20, 20, 0.8, 20, StandardRobot))
        times2.append(run_simulation(num_robots, 1.0, 20, 20, 0.8, 20, RandomWalkRobot))
    print("\r            ")
    
    plt.plot(num_robot_range, times1, label='StandardRobot')
    plt.plot(num_robot_range, times2, label='RandomWalkRobot')
    plt.title(title)
    plt.legend()
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.show()

plot_needed_time('Time It Takes 50 Robots To Clean 80% Of A Room' ,' Number of Robots' ,'Time-steps', range(1, 51))

plot_needed_time('Time It Takes 10 Robots To Clean 80% Of A Room' ,' Number of Robots' ,'Time-steps', range(1, 11))
```

```python
def plot_cleaning_time(title, x_label, y_label):
    """
    What information does the plot produced by this function tell you?
    """
    aspect_ratios = []
    times1 = []
    times2 = []
    for width in [10, 20, 25, 50]:
        height = 300//width
        print("Plotting cleaning time for a room of width:", width, "by height:", height)
        aspect_ratios.append(float(width) / height)
        times1.append(run_simulation(2, 1.0, width, height, 0.8, 200, StandardRobot))
        times2.append(run_simulation(2, 1.0, width, height, 0.8, 200, RandomWalkRobot))
    plt.plot(aspect_ratios, times1, label='StandardRobot')
    plt.plot(aspect_ratios, times2, label='RandomWalkRobot')
    plt.title(title)
    plt.legend()
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.show()


plot_cleaning_time('Time It Takes Two Robots To Clean 80% Of Variously Shaped Rooms' ,'Aspect Ratio' ,'Time-steps')
```
